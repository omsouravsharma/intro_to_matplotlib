## Matplot Example worflow

```python
# 0. Import matplotlib

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Set up plot

fig, ax = plt.subplots(figsize = (10, 10)) # [width, height]

# 3. Plot data

ax.plot(x,y)

#4. Customized plot

ax.set(title="Sample Plot", 
      xlabel = "x-axis", 
      ylabel = "y-axis")

# 5. Save 
fig.savefig("images/sample-plot.png")
```

### Subplots within figure 


```python
#Option -1 

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                            ncols=2, 
                                            figsize=(10,5))

#Plot to each different axis: 

ax1.plot(x,x/1);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option 2: 
fig, ax = plt.subplots(nrows=2, 
                      ncols=2,
                      figsize=(10, 5))

# Plot to each index: 

ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# pyplot vs matplotlib
* Quickly - Pyplot
* Advance - OO method

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
#Pyplot
over_50.plot(kind= "scatter", 
            x = "age", 
            y= "chol",
            c="target");
```

```python
## OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", 
            x ="age",
            y= "chol",
            c= "target",
            ax =ax)
# ax.set_xlim([45, 100])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
#Plot the data
scatter = ax.scatter( x=over_50["age"], 
                    y = over_50["chol"],
                    c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age", ylabel="Cholestrol")

# Legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50
```

```python
#Subplot of thalach

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1,figsize=(10,10))
scatter = ax0.scatter(x =over_50["age"], 
                     y =over_50["chol"],
                     c= over_50["target"])

ax0.set(title="Heart Disease and Cholestrol Level", 
       ylabel = "Cholestrol")

ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(over_50["chol"].mean(),
           linestyle="--")

ax1.scatter(x=over_50["age"],
           y=over_50["thalach"],
           c=over_50["target"])
ax1.set(title="Heart Disease and MAx Heart Disease",
       xlabel="Age",
       ylabel="MAx Heart Rate")
ax1.legend(*scatter.legend_elements(), title="target")
ax1.axhline(over_50["thalach"].mean(), linestyle="--")

# Add title to figure

fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold") 
```

```python
plt.style.use('seaborn-whitegrid')
#Subplot of thalach

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1,figsize=(10,10))
scatter = ax0.scatter(x =over_50["age"], 
                     y =over_50["chol"],
                     c= over_50["target"],
                     cmap="winter")

ax0.set(title="Heart Disease and Cholestrol Level", 
       ylabel = "Cholestrol")

ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(over_50["chol"].mean(),
           linestyle="--")
ax0.set_xlim([50,80])

scatter_2 = ax1.scatter(x=over_50["age"],
           y=over_50["thalach"],
           c=over_50["target"],
                       cmap="winter")
ax1.set(title="Heart Disease and MAx Heart Disease",
       xlabel="Age",
       ylabel="Max Heart Rate")
ax1.legend(*scatter_2.legend_elements(), title="target")
ax1.axhline(over_50["thalach"].mean(), linestyle="--")
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add title to figure

fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold") 
```
